```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dft = dft.rename(columns={'Marital_Status': 'Marital'})
dft
```

```python
dft['Marital'] = dft['Marital'].replace({0: 'single', 1: 'married'})
dft
```

```python
dft['Gender'] = dft['Gender'].replace({'F': 'Female', 'M': 'Male'})
dft
```

```python
plt.figure(facecolor='pink',figsize=(28,9)  )
plt.title('Sales Data',fontsize=20)
plt.subplot(2,3,1)
a1 = sns.countplot(x='Gender',data=dft,facecolor='yellow')
plt.gca().set_facecolor('blue')
for i in a1.containers:
 a1.bar_label(i)
#  plt.figure(facecolor='yellow')
plt.subplot(2,3,2)
a1=sns.barplot(x='Gender',y='Amount',data=dft,facecolor='pink',estimator='sum',errorbar=None)
plt.gca().set_facecolor('blue')
for i in a1.containers:
 a1.bar_label(i)
 plt.subplot(2,3,3)
sns.barplot(y='Age Group',x='Amount',data=dft,facecolor='pink',estimator='sum',errorbar=None)
plt.gca().set_facecolor('blue')
# for i in a1.containers:
#  a1.bar_label(i)
plt.subplot(2,3,4)
sns.barplot(y='State',x='Amount',data=dft,facecolor='pink',estimator='sum',errorbar=None)
plt.gca().set_facecolor('blue')

plt.subplot(2,3,5)
sns.barplot(y='Occupation',x='Amount',data=dft,facecolor='pink',estimator='sum',errorbar=None)
plt.gca().set_facecolor('blue')

plt.subplot(2,3,6)
sns.barplot(y='Product_Category',x='Amount',data=dft,facecolor='pink',estimator='sum', errorbar=None)
plt.gca().set_facecolor('blue')
plt.tight_layout()
plt.show()
```

## Conclusion

Based on the analysis of the sales data, we can draw the following conclusions:

*   **Gender Distribution:** The sales data shows a higher number of female customers compared to male customers.
*   **Sales by Gender:** The higher number of female customers, the total amount spent by each gender is relatively similar. This suggests that while there are more female shoppers, male shoppers tend to spend more per transaction.
*   **Sales by Age Group:** The age group with the highest amount of sales is the 26-35 age group. and the lower amount of sales is 0-17 age group.
*   **Sales by State:**  The top performing states in terms of sales amount is Uttarpradesh and lower amount of is Telengana.
*   **Sales by Occupation:**The top performing occupations in terms of sales amount is It sector.
*   **Sales by Product Category:** Top performing product categories in terms of sales amount is Food.
*   **Sales by Marital Status:**  Compare the sales amount between single and married individuals.

These findings can be used to inform marketing strategies and target specific demographics and product categories to maximize sales.
